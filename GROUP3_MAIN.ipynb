```python
import pandas as pd
```

```python

df = pd.read_excel(r"data/Edinburgh-daytime.xlsx",sheet_name='1112', skiprows=3, header=None)
print(df) 
```

```python
#for future use, to automate the pipeline and consider other sheets we would take the year and month from the sheet names. 
#However, as we are focusing on one sheet only for now, we will hard code it. 

year = 2011
month = 12

print(year)
print(month)
```

```python
#Dictionary assigned to a variable called Edinburgh
EDINBURGH_COLUMNS = {
    'date':pd.Timestamp,
    'sunrise': str,
    'sunset': str,
    'daylength': str,
    'day_length_diff': str,
    'astro_twilight_start': str,
    'astro_twilight_end': str,
    'nautical_twilight_start': str,
    'nautical_twilight_end': str,
    'civil_twilight_start': str,
    'civil_twilight_end': str,
    'solar_noon_time': str,
    'solar_noon_distance':float
}
```

```python
#Let's reset the index - dropping the second header row, as it's not needed
#Now the row count starts from the actual data

#df = df.drop([0, 1], ).reset_index(drop=True)
#print(df) 
```

```python
#Fix the columns

#df.columns = list(EDINBURGH_COLUMNS.keys())
df.columns = list(EDINBURGH_COLUMNS)
print(df.head())



```

EDINBURGH FILE - DAYTIME DATA

```python
df_clean = df[['date','sunrise','sunset','daylength','solar_noon_time']]
print(df_clean.head())

df_clean.to_csv("data/clean_daytime_data.csv", index=False) 
```

```python
#read the new csv file. 
df_ED = pd.read_csv(r"data/clean_daytime_data.csv")

df_ED.head()
```

```python
#Quick test for one column
#df['sunrise'] = df['sunrise'].str[:5] 
#print(df['sunrise'] )
```

```python
#remove extra, unnecesary information

for col in ['sunrise', 'sunset', 'solar_noon_time']:
            df_ED[col] = df_ED[col].str[:5] 
```

```python
df_ED.head()
```

```python
#Add date values - We know that this file is for December 2011
import datetime

df_ED['date'] = pd.to_datetime(df_ED['date'].apply(lambda x: datetime.date(year, month, x))) 
print(df_ED['date'])
```

Remove empty rows is key cleaning step, but we are not doing this as we have selected a sheet without them and we wouldn't be able to test this. 

Strathspey - WEATHER DATA

```python
df = pd.read_excel(r"data/Strathspey-weather.xlsx",sheet_name='1112', skiprows=6, header=None)
print(df) 
```

```python
STRATHSPEY_COLUMNS = {
    'date':pd.Timestamp,
    'temp_mean': str,
    'temp_min': str,
    'temp_max': str,
    'rain_mm': str,
    'pressure_early': int,
    'pressure_late': int,
    'wind_mean':float,
    'wind_max':float,
    'wind_dir':str,
    'sun_hours': float
}
```

```python
#rename columns

df.columns = list(STRATHSPEY_COLUMNS)
print(df.head())
```

```python
#select columns to keep 

columns_to_keep = ['date', 'temp_mean', 'temp_min', 'temp_max']

df_clean = df[columns_to_keep]

# Save to new CSV
df_clean.to_csv("data/clean_temperature_data.csv", index=False) 
```

```python
#read new, clean csv file

df_ST = pd.read_csv(r"data/clean_temperature_data.csv", encoding = 'utf-8')
print(df_ST) 
```

```python
print(df_ST.dtypes)
```

```python
df_ST['temp_min']  = df_ST['temp_min'] .astype(str).str.replace(r"\s+", "", regex=True).astype(float)

print(df_ST)

```

```python
print(df_ST.dtypes)
```

```python
print(df_ST)

```

```python
#Add date values
df_ST['date'] = pd.to_datetime(df_ST['date'].apply(lambda x: datetime.date(year, month, x))) 

#print(df['date'])

print(df_ST)

```

MERGE BOTH DATA FRAMES

```python
merged = pd.merge(df_ED, df_ST, on="date", how="inner") 

print(merged)
```

```python
#Save merge dataset to a new csv file
merged.to_csv("output/merged.csv", index=False) 
```

DATA VALIDATIONS
- row count
- duplicate dates
- missing values when merging


```python

#Test - file count validation. Checking that row counts of all three files match each other. 

df_merged = pd.read_csv("output/merged.csv")

assert len(df_ED) == len(df_ST), (
    f"Input row mismatch: file1 has {len(df_ED)}, file2 has {len(df_ST)}"
)

assert len(df_merged) == len(df_ED), (
    f"Merged row mismatch: expected {len(df_ED)}, got {len(df_merged)}"
)

print("✓ Row count test passed") 

```
